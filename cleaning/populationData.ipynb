```python
# Modules to run code
import csv
import pandas as pd 
```

```python
# Import .csdv files from Resources folder
PopData = pd.read_csv('../data/csvpopData_2.csv')
```

```python
# Showing dataframe

PopData
```

```python
# Deleting columns from Dataframe
cleaned = PopData.drop(columns=['rank','Growth','Pop2018','Pop2010','growthSince2010', 'Percent'])
```

```python
# reading in csv with lat long
latdf = pd.read_csv('../data/Lat_long_states.csv') 
latdf
```

```python
# merging pop df and lat long df on state 
merged_pop_df = pd.merge(cleaned,latdf, on='State', how='left')
merged_pop_df
```

```python
# Export to csv
merged_pop_df.to_csv("../data/cleanedPop_lat_Data.csv", index=False)
```
